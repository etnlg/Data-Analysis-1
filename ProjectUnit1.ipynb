Next, we're going to compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

So we're facing a conditionnal probability:
Let the event A be the event that a commune has an elevation over 2000, and let the event B be the event that a commune is from the canton valais. Therefore, we have the following equation:

P(B⎢A)=P(AB) / P(A)

So, first, let's find P(A), which is the probability that a commune has an elevation over 2000

```python
high_altitude=df[df['Elevation']>2000]
P_A=len(high_altitude['Elevation'].tolist())/len(df['Elevation'].tolist())
P_A
```

Now that we found P(A), let's compute P(AB), which is when both event occur

```python
high_valais=high_altitude[high_altitude['Canton']=='VS']
P_AB=len(high_valais['Canton'].tolist())/len(df['Canton'].tolist())
P_AB
```

Now, we can replace with the values

```python
Probability=P_AB/P_A
Probability
```

Therefore, the probability is 0.51

Now, we're going to define a data frame matrix whose rows correspond to communes and the columns to the cantons

```python
df.reset_index(inplace=True)
```

```python
columns=df['Canton'].unique().tolist()
rows=df['Commune']
n_columns=len(columns)
n_rows=len(rows)
matrix=np.zeros((n_rows,n_columns))
for i in range (n_rows):
    canton_row=df['Canton'][i]
    
    for j in range (n_columns):
        canton_column=columns[j]
        
        if canton_row==canton_column:
            matrix[i,j]=1

DataFrame_matrix=pd.DataFrame(data=matrix,index=rows,columns=columns)
DataFrame_matrix
```
